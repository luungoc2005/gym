```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import math
```

```python
all_tickers = ['U11', 'BS6', 'A17U', 'G13', 'O39', 'Z74', 'C61U', 'C31', 'C38U', 'D05']
fig = plt.figure()
plt_rows = math.ceil(len(all_tickers) / 2) + 3
net_worth_ax = plt.subplot2grid(
    (plt_rows, 2), (0, 0), 
    rowspan=2, colspan=2
)
tickers_plt = []
for ticker_ix, ticker in enumerate(all_tickers):
    ticker_x = ticker_ix % 2
    ticker_y = ticker_ix // 2 + 2
    ticker_plt = plt.subplot2grid(
        (plt_rows, 2), (ticker_y, ticker_x), 
        rowspan=1, colspan=1, 
        sharex=net_worth_ax
    )
    ticker_plt.set_title(ticker)
    tickers_plt.append(ticker_plt)

plt.subplots_adjust(left=0.12, bottom=0.25, right=0.9, 
      top=0.9, wspace=0, hspace=0)

plt.show(block=False)
```
